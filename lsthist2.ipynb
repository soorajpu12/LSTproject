```
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import rioxarray as rxr

# Prettier plotting with seaborn
sns.set(font_scale=1.5, style="whitegrid")
```

```
img1 = os.path.join('/content/drive/MyDrive/code_editor1/lst1.tif')
```

```
img2 = rxr.open_rasterio(img1)
img2.shape
```

```
f, ax = plt.subplots(figsize=(10, 6))

img2.plot.hist(ax=ax,
       color="purple",bins=100)

ax.set(title="Distribution of mean LST values",
       xlabel='mean LST (°C)',
       ylabel='Frequency')
plt.show()
```
